Did you know that prime numbers, those special numbers that can only be divided by 1 and themselves, can be used to keep our information secure? It may seem surprising, but prime numbers play a vital role in the field of encryption, where messages are transformed into secret codes to prevent unauthorized access. 

This homework explores the idea of how exactly to encrypt and why decrypting with the correct "password" is easy while decrypting with the wrong "password" is difficult. 

1. Write a class called "PrimeGen" that generate two consecutive prime numbers that are greater than the input n. For example, ps = PrimeGen(10) should output [11,13]

```python
class PrimeGen:
    def __init__(self, n):
        foundWhich = 0
        copy = n
        while (foundWhich != 2):
            copy += 1
            found1 = True
            for i in range(2, (int)(copy / 2)):
                if copy % i == 0: 
                    found1 = False
                    break
            
            if (found1 and foundWhich == 0): 
                self.x1 = copy
                foundWhich += 1
                continue
            if (found1 and foundWhich == 1):
                self.x2 = copy
                foundWhich += 1
    
    def __str__(self):
        return "[" + str(self.x1) + ", " + str(self.x2) + "]"
        

ps = PrimeGen(10)
```

```python
print(ps)
```

2. Write a subclass called "Encrypt". In the subclass, create a method called 'encrypt()' that multiplies the two consecutive prime numbers generated by "PrimeGen". For example, msg = Encrypt(10).encrypt() should output 143. 

```python
class Encrypt(PrimeGen):
    def encrypt(self):
        return self.x1 * self.x2
```

```python
msg = Encrypt(10).encrypt()
msg
```

3. Write a function called "decrypt" that given a number n and a key k, return n/k if k is indeed a factor of n. If k is not a factor of n, return "wrong key".For example, decrypt(143,11) should give 13, decrypt(22,2) should give 11, and decrypt(25,3) should give "wrong key"

```python
def decrypt(number, key):
    if number % key == 0: return (int)(number / key)
    return "wrong key"
```

```python
decrypt(143, 11)
```

4. Write a function called "hack" that tries to hack an encryption. Namely, find two factors of a number by brute force. 

```python
def hack(number):
    for i in range(2, (int)(number ** 0.5) + 1):
        if number % i == 0: return "[" + str(i) + ", " + str((int)(number / i)) + "]"
    return "Prime Number"
```

```python
hack(143)
```
